```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha', 'Zoya']

print("Идеальные пары:" if len(boys) == len(girls) else "Внимание, кто-то может остаться без пары!")
if len(boys) == len(girls):
    for boy, girl in zip(sorted(boys), sorted(girls)):
        print(f"{boy} и {girl}")
```
